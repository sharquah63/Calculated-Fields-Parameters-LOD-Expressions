Theoretical Questions :

Q1. What is a Calculated Field and why is it important in business dashboards?

-> A calculated field is a new field created from existing data using a formula or logic. It is important because it helps turn raw data into meaningful metrics such as profit margin, growth rate, delivery status, or performance categories. In business dashboards, it allows analysts to create custom insights that are not directly available in the dataset.

Q2.  Why can’t parameters change visuals directly in Tableau?

-> Parameters do not directly filter or modify data by themselves. They are user-controlled values that must be connected to calculated fields, filters, or reference lines to affect visuals. So, a parameter only changes the visual indirectly through a calculation or rule.

Q3. When should EXCLUDE LOD be used?

-> EXCLUDE LOD should be used when you want to calculate a value at a higher level of aggregation by ignoring one or more dimensions currently in the view. It is useful when you need to remove detail from the current visualization and see a broader summary. For example, you can compare category-level performance while excluding product-level detail.

Q4. Difference between FIXED and INCLUDE LOD expressions?

-> FIXED LOD calculates a value at a specified level of detail regardless of what dimensions are in the view. INCLUDE LOD adds extra detail to the current view before calculating the result, so it depends on the dimensions present in the visualization. In short, FIXED ignores the view level, while INCLUDE expands it.
